```python
#En al columna a de frame llamamos a los no nulos
cframe=frame[frame.a.notnull()]
cframe
```

```python
#Agregaoms la columna os, colocando windows o no windows respectivamente
cframe['os'] = np.where(cframe['a'].str.contains('Windows'),'Windows', 'Not Windows')
```

```python
#Extraemos los 11 primeros
cframe[:11]
```

```python
#Extraemos la columna "OS"
cframe["os"]
```

```python
#Agrupamos por "tz" y "os"
by_tz_os =cframe.groupby(['tz', 'os'])
cframe.groupby(['tz', 'os']).sum()
```

```python
#Contamos por tz por windwos o not windows
agg_counts = by_tz_os.size().unstack().fillna(0)
```

```python
#Ejemplo de unstack
by_tz_os.size().unstack().fillna(0)
```

```python
#Extraemos los 10 primeros de "agg_counts"
agg_counts[:10]
```

```python
#Extraemos los indices de menor a mayor
indexer = agg_counts.sum(1).argsort()
```

```python
#Ejemplo de argsort
x= np.array([5, 3,1,2, 6])
x.argsort()
```

```python
#Que hace agg_counts.sum(1)
list(agg_counts.sum(1).argsort())
```

```python
#Comprobacion de indices
agg_counts.sum(1)[-10:]
```

```python
indexer[-10:]
```

```python
count_subset = agg_counts.take(indexer[-10:])
```

```python
count_subset
```

```python
#Reordenando la data
count_subset=count_subset.stack()
count_subset
```

```python
#Colocando nombre total a la cantidades
count_subset.name="total"
```

```python
#Reseteamos el index
count_subset = count_subset.reset_index()
```

```python
#Extraemos los 10 primeros
count_subset[:10]
```

```python
import seaborn as sns
```

```python
#Graficamos en X el total y en Y los paises/ciudades
sns.barplot(x='total', y='tz', hue='os', data=count_subset)
```

```python
#Problema--> no se indentifica los numeros pequeños 
```

```python
#Solucion --> windows + not windows = 1
```

```python
#Agrega nueva columna diviendo ("windowns/not windows") entre la suma de ambos en su tz
def norm_total(group):
 group['normed_total'] = group.total / group.total.sum()
 return group
```

```python
#Agrupamos por "tz" y aplicamos la funcion norm_total
results = count_subset.groupby('tz').apply(norm_total)
results
```

```python
#Ya casi terminamos
```

```python
#Graficamos con x en porcentaje, Y en con "tz"
sns.barplot(x="normed_total", y='tz', hue='os', data=results)
```
